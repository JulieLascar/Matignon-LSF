```python
import spacy
spacy.prefer_gpu()
nlp = spacy.load("fr_dep_news_trf")
from spacy.language import Language
```

```python
def get_dict_vtt(input):
    with open(input,encoding="utf-8") as f:
        lines = f.readlines()

    dict_sub = {}
    i = 0
    j = 0  # Déplacez la déclaration de 'j' en dehors de la boucle for

    while j < len(lines):  # Utilisez une boucle while plutôt qu'une boucle for
        element = lines[j]
        if element.startswith("00:") or element.startswith("01:"):
            # Extraire le temps de début et de fin
            timing_line = element.strip().split(' --> ')
            start_time, end_time = timing_line

            text = ""
            while j + 1 < len(lines) and not lines[j + 1].startswith("0"):
                j += 1
                content = lines[j].replace("[INAUDIBLE]","")
                text = text + " " + content.strip()

            dict_sub[i] = {'start': start_time, 'end': end_time, 'text': text.strip()}
            i += 1

        j += 1

    return dict_sub
```

```python
def create_vtt_file(dictionary, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as file:
        file.write("WEBVTT\n\n")

        for key, v in dictionary.items():
            for kk,vv in v.items():
                if kk == 'start':
                    start_time = vv
                if kk == 'end':
                    end_time = vv
                if kk == 'text':
                    text = vv

            vtt_entry = f"{start_time} --> {end_time}\n{text}\n\n"
            file.write(vtt_entry)
```
