### plot the images of the 15 signers

```python
# plot the images of the 15 signers (just one per signer)
signers_nb = 15

fig = plt.figure(figsize=(15, 12))
columns = 5
rows = 3

ax = []
cpt = 1
for i in range(1,signers_nb+1):
    
    img = cv2.imread(f'signers_detection/signers_images/img{i}.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    ax.append( fig.add_subplot(rows, columns, cpt) )
    ax[-1].set_title(str(i))
    plt.axis('off')
    plt.imshow(img)

    cpt +=1
    
plt.show()
fig.savefig('signers_detection/signers.png', bbox_inches='tight')
```
